```python
import pandas as pd
```

```python
df = pd.read_csv('shifted4.csv')
labels = pd.read_csv('labels4.csv')
```

```python
print(df.info())
df.head()
```

```python
df.drop(df.columns[[10]], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df.iloc[:,0].value_counts())
```

```python
l=['X-RAY DIFFRACTION','SOLUTION NMR','ELECTRON MICROSCOPY','SOLID-STATE NMR','ELECTRON CRYSTALLOGRAPHY','NEUTRON DIFFRACTION','FIBER DIFFRACTION','SOLUTION SCATTERING','POWDER DIFFRACTION','THEORETICAL MODEL','EPR','INFRARED SPECTROSCOPY','ELECTRON DIFFRACTION','CRYO-ELECTRON MICROSCOPY']                      

for col in l:
    for i in range(len(df)):
        if(col in df.iloc[i,0]):
            df.loc[i,col]=1
        else:
            df.loc[i,col]=0

df.head()
```

```python
df.to_csv('encode.csv',index=False)
```

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

encoded_data = encoder.fit_transform(df[['macromoleculeType']])

encoded_df = pd.DataFrame(encoded_data.toarray(), columns=[f"{col}_{val}" for col, val in zip(df['macromoleculeType'].unique(), encoder.get_feature_names_out())])

print("\nOne-Hot Encoded DataFrame:")
encoded_df.head()
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.info()
```

```python
columns_to_drop = ['experimentalTechnique', 'macromoleculeType','crystallizationMethod']
df = df.drop(columns=columns_to_drop)
```

```python
df.isnull().sum()
```

```python

df['resolution'] = df['resolution'].fillna(df['resolution'].mean())
df['crystallizationTempK'] = df['crystallizationTempK'].fillna(df['crystallizationTempK'].mean())
df['densityMatthews'] = df['densityMatthews'].fillna(df['densityMatthews'].mean())
df['densityPercentSol'] = df['densityPercentSol'].fillna(df['densityPercentSol'].mean())
df['phValue'] = df['phValue'].fillna(df['phValue'].mean())

```

```python
df.isnull().sum()
```

```python
df.to_csv('finalTrain.csv',index=False)
```
